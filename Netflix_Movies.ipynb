```python
# Create the years and durations lists(Yıl ve film süreleri içeren listeleri yarat)
years = [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
durations = [103,101,99,100,100,95,95,96,93,90]

# Create a dictionary with the two lists(Bu iki listeyle bir sözlük yarat)
movie_dict = {
             "years":years,
             "durations":durations}

# Print the dictionary(Sözlüğü yazdır)
movie_dict
```

```python
# Import pandas(Pandas kütüphanesini içe aktar)
import pandas as pd

# Create a DataFrame from the dictionary(Oluşturduğumuz sözlükle bir Dataframe yarat)
durations_df = pd.DataFrame(movie_dict)

# Print the DataFrame(DataFrame'i yazdır)
print(durations_df)
```

```python
# Import matplotlib.pyplot(Matplotlib kütüphanesini içe aktar)
import matplotlib.pyplot as plt
fig = plt.figure()

# Draw a line plot of release_years and durations(release_year ve durations kolonlarıyla çizgi grafiği oluştur)
plt.plot(durations_df.years,durations_df.durations)

# Create a title/Başlık oluşturma
plt.title("Netflix Movie Durations 2011-2020")

# Show the plot(Grafiği yazdırma)
plt.show()
```

```python
# Read in the CSV as a DataFrame(CSV dosyasını Dataframe ile okuma)
netflix_df = pd.read_csv("netflix_data.csv")

# Print the first five rows of the DataFrame(Data Frame'in ilk 5 satırını yazdırma)
netflix_df.head()
```

```python
# Subset the DataFrame for type "Movie"(Type kolonunda sadece "movie" olanları alma)
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']

# Select only the columns of interest(Üstünde çalışacağımız kolanları çekmek)
netflix_movies_col_subset = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]

# Print the first five rows of the new DataFrame(İlk 5 satırı yazdırmak)
netflix_movies_col_subset.iloc[:5]
```

```python
# Create a figure and increase the figure size
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus year(Dağılım grafiği oluşturma iki kolan arasında)
plt.scatter(netflix_movies_col_subset.release_year,netflix_movies_col_subset.duration)

# Create a title(Başlık oluşturmak)
plt.title("Movie Duration by Year of Release")
# Show the plot(Grafiği çizdirmek)
plt.show()
```

```python
# Filter for durations shorter than 60 minutes(Süresi 60 dk'dan az olan filmleri çekmek)
short_movies = netflix_movies_col_subset[netflix_movies_col_subset["duration"]<60]

# Print the first 20 rows of short_movies(Aldığımız kısa filmlerin 20 tanesini yazdırmak)
short_movies[:20]
```

```python
# Define an empty list(renkleri ekleyebilceğimiz boş bir liste oluşturma)
colors = []

# Iterate over rows of netflix_movies_col_subset(Dataframe'den kolon ve indexleri almak için for döngüsü kullanmak)
for i,j in netflix_movies_col_subset.iterrows() :
    if j.genre=="Children" :
        colors.append("red")
    elif j.genre=="Documentaries" :
        colors.append("blue")
    elif j.genre=="Stand-Up" :
        colors.append("green")
    else:
        colors.append("black")

# Inspect the first 10 values in your list (Oluşturduğumuz listenin ilk 10 elemanına bakmak)
colors[:11]
```

```python
# Set the figure style and initalize a new figure(Oluşturcağımız tablonun şeklini ve stilini ayarlamak)
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus release_year(Renk listesini kullanarak dağılım grafiği oluşturmak)
plt.scatter(netflix_movies_col_subset.release_year,netflix_movies_col_subset.duration,color=colors)

# Create a title and axis labels(Başlık ve eksenlere isim vermek)
plt.title("Movie duration by year of release")
plt.xlabel("Release year")
plt.ylabel("Duration(min)")

# Show the plot
plt.show
```
